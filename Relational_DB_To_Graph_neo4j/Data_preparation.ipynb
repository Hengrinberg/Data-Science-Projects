```python
previous_application = pd.read_csv('C:/Users/heng/Desktop/Yael - Active Projects/credit default risk - kaggle competition/previous_application.csv')
```

```python
import math
# I looked at the data and saw that there is almost 100% correlation in the number of missing values in some columns:
# AMT_ANNUITY-CNT_PAYMENT
# RATE_INTEREST_PRIMARY-RATE_INTEREST_PRIVILEGED
# AMT_DOWN_PAYMENT-RATE_DOWN_PAYMENT
# DAYS_FIRST_DRAWING-DAYS_FIRST_DUE-DAYS_LAST_DUE_1ST_VERSION-DAYS_LAST_DUE-DAYS_TERMINATION-NFLAG_INSURED_ON_APPROVAL

# I will generate 4 more features to the table which checking if each combination exists or not
first_combination = []
second_combination = []
third_combination = []
forth_combination = []
for i in range(previous_application.shape[0]):
    if math.isnan(previous_application.loc[i,'AMT_ANNUITY']) == True and math.isnan(previous_application.loc[i,'CNT_PAYMENT']) == True:
        first_combination.append(1)
    if math.isnan(previous_application.loc[i,'AMT_ANNUITY']) == False or math.isnan(previous_application.loc[i,'CNT_PAYMENT']) == False:
        first_combination.append(0)
        
    if math.isnan(previous_application.loc[i,'RATE_INTEREST_PRIMARY']) == True and math.isnan(previous_application.loc[i,'RATE_INTEREST_PRIVILEGED']) == True:
        second_combination.append(1)
    if math.isnan(previous_application.loc[i,'RATE_INTEREST_PRIMARY']) == False or math.isnan(previous_application.loc[i,'RATE_INTEREST_PRIVILEGED']) == False:
        second_combination.append(0)
        
    if math.isnan(previous_application.loc[i,'AMT_DOWN_PAYMENT']) == True and math.isnan(previous_application.loc[i,'RATE_DOWN_PAYMENT']) == True:
        third_combination.append(1)
    if math.isnan(previous_application.loc[i,'AMT_DOWN_PAYMENT']) == False or math.isnan(previous_application.loc[i,'RATE_DOWN_PAYMENT']) == False:
        third_combination.append(0)
        
    if math.isnan(previous_application.loc[i,'DAYS_FIRST_DRAWING']) == True and math.isnan(previous_application.loc[i,'DAYS_FIRST_DUE']) == True and math.isnan(previous_application.loc[i,'DAYS_LAST_DUE_1ST_VERSION']) == True and math.isnan(previous_application.loc[i,'DAYS_LAST_DUE']) == True and math.isnan(previous_application.loc[i,'DAYS_TERMINATION']) == True and math.isnan(previous_application.loc[i,'NFLAG_INSURED_ON_APPROVAL']) == True:
        forth_combination.append(1)
    if math.isnan(previous_application.loc[i,'DAYS_FIRST_DRAWING']) == False or math.isnan(previous_application.loc[i,'DAYS_FIRST_DUE']) == False or math.isnan(previous_application.loc[i,'DAYS_LAST_DUE_1ST_VERSION']) == False or math.isnan(previous_application.loc[i,'DAYS_LAST_DUE']) == False or math.isnan(previous_application.loc[i,'DAYS_TERMINATION']) == False or math.isnan(previous_application.loc[i,'NFLAG_INSURED_ON_APPROVAL']) == False:
        forth_combination.append(0)

previous_application['first_combination'] = first_combination
previous_application['second_combination'] = second_combination
previous_application['third_combination'] = third_combination
previous_application['forth_combination'] = forth_combination
```
